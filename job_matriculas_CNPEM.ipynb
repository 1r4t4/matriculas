## Situação: Solicitaram-me o número de matrículas na rede escolar, a partir do ensino fundamental, de todas cidades distantes até 130 km do município referência.
## Para levantamento dos dados de todo o estado de SP foi importada uma tabela do banco QEdu. Para definir as cidades distantes do município referência foi utilizada a base contida no domínio "emsampa.com.br".

# Código desenvolvido em Phyton no Colab

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados_qedu.csv')
```

```python
dados.head()
```

```python
Cidade = dados['Cidade']
Cidade.head()
```

```python
Matriculas = dados['Total de matrículas']
Matriculas
```

```python
matriculas = pd.DataFrame({'Cidade': Cidade, 'Matrículas': Matriculas})
matriculas.head()
```

```python
soma_das_matriculas = sum(matriculas['Matrículas'])
print(soma_das_matriculas)
```

```python
cidades_130 = pd.read_csv('cidades_130.csv')
cidades_130.head()
```

```python
lista_cidades_130 = cidades_130['Cidade'].tolist()
lista_cidades_130
```

```python
selecao_130 = matriculas['Cidade'].isin(lista_cidades_130)
cidades_selecionadas_130 = matriculas[selecao_130]
```

```python
cidades_selecionadas_130.head()
```

```python
soma_das_matriculas_130 = sum(cidades_selecionadas_130['Matrículas'])
print(soma_das_matriculas_130)
```

Existem 833033 matrículas nos municípios até 130 km.
